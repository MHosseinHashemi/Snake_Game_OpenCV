```python
class Snake:
    def __init__(self):
        self.preHead = 0,0 # lastest Head point
        self.joints = [] # Joints of snake body
        self.lengths = [] # Distance between each joint
        self.currentLen = 0 # Snake current length
        self.lenThresh = 180 # Game Goal
        # Bonus Location
        self.Bonus_Loc = 0, 0
        # Just right after initializing init Location
        self.random_Bonus()
        self.score = 0
        self.lenProg = 0
        self.Goal = 500
    
    
    def random_Bonus(self):
        self.Bonus_Loc = random.randint(200,900), random.randint(200,600) 
    
    def update(self, img, head):
        px, py = self.preHead
        nx, ny = head
        
        # Append them to the joints
        self.joints.append([nx, ny])
        
        # The math.hypot() method returns the Euclidean norm. 
        # The Euclidian norm is the distance from the origin to the coordinates given
        distance = math.hypot(nx-px, ny-py)
        self.lengths.append(distance)
        self.currentLen += distance
        self.preHead = nx, ny
        
        # Drawing the snake
        for idx, joint in enumerate(self.joints):
            if idx != 0:
                cv2.line(img, self.joints[idx-1], self.joints[idx], (120,0,0), 15)
        cv2.circle(img, self.joints[-1], 18, (0,0,200), cv2.FILLED)
        
        # Drawing the Bonus
        rx, ry = self.Bonus_Loc
        rx = int((rx/2))
        ry = int((ry/2))
        cv2.circle(img, (rx,ry), 25, (0,180,0), cv2.FILLED)
        
        # Length Handling
        if self.joints:    
            if self.currentLen >= self.lenThresh:
                for i, length in enumerate(self.lengths):
                        self.currentLen -= length
                        self.lengths.pop(i)
                        self.joints.pop(i)
                        # The loop Way out
                        if self.currentLen < self.lenThresh:
                            break 
        
        
        # Prize Handling
        if rx-40 < nx < rx+40 and ry-40 < ny < ry+40:
            self.score += 1
            self.random_Bonus()
            rx, ry = self.Bonus_Loc
            rx = int((rx/2))
            ry = int((ry/2))
            cv2.circle(img, (rx,ry), 25, (0,180,0), cv2.FILLED)
            # Now we need to add to the size of the snake
            temp = self.lenThresh
            self.lenThresh += self.lenThresh*0.05
            self.lenProg = 100 - int(((self.Goal - self.lenThresh)/self.Goal)*100)
            
        
        if self.lenProg >= 100:
            cv2.putText(img, ('You Won !'),
                tuple(np.multiply((0.22,0.45), [1280, 720]).astype(int)),
                cv2.FONT_HERSHEY_SIMPLEX, 5, (0,255,0), 8, cv2.LINE_AA)
            
        elif self.lenProg < 100:        
            cv2.putText(img, (f'Score : {self.score}'),
                        tuple(np.multiply((0.77,0.07), [1280, 720]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0,165,255), 2, cv2.LINE_AA)

            cv2.putText(img, (f'Pregress : % {self.lenProg}'),
                        tuple(np.multiply((0.77,0.12), [1280, 720]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0,165,255), 2, cv2.LINE_AA)
            

            
        return img
```

```python
"""Latest Version"""
# Creating the Cap Object
cap = cv2.VideoCapture(0)
cap.set(3, 1280)
cap.set(4, 720)

hand_map_drawing = mp.solutions.drawing_utils
map_hands = mp.solutions.hands

with map_hands.Hands(min_detection_confidence = 0.5, min_tracking_confidence = 0.5, max_num_hands = 1 ) as hands:
    
    while(cap.isOpened()):
        ret, frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Flip the Frame to make it real(Mirror Effect of Webcam)
        image.flags.writeable = False
#         image = cv2.flip(image, 1)
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    
        if results.multi_hand_landmarks:               
            
            for hand in results.multi_hand_landmarks:
                Index_Tip = int(1280*(hand.landmark[map_hands.HandLandmark.INDEX_FINGER_TIP.value].x)), int(720*(hand.landmark[map_hands.HandLandmark.INDEX_FINGER_TIP.value].y))
                img = game.update(image, Index_Tip)
            
    
    
        cv2.putText(image, ('Press Q to Quit'),
                    tuple(np.multiply((0.01,0.07), [1280, 720]).astype(int)),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv2.LINE_AA)
        
        cv2.imshow('Snake Game', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
            
cap.release()
cv2.destroyAllWindows()
```
